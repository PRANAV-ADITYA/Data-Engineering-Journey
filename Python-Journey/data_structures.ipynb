```python
[x**2 for x in range(10)] 
```

```python
square = [ num ** 2 for num in range(10) if(num%2==0)]
square
```

```python
# Nested List Comprehension

lst1 = [1,2,3,4]
lst2 = ['a','b','c','d']
result = [(i,j) for i in lst1 for j in lst2]
result
```

# Packing and Unpacking of tuple

```python
packed_tuple = 1, "Hello" , 3.14, 2 ,3 , 5 ,6,7,8  ## Packing
print(packed_tuple)

```

```python
a , b , c  = packed_tuple ## Unpacking
print(a)
print(b)
print(c)
```

```python
a , b , *c  = packed_tuple
print(a)
print(b)
print(*c) ## Here it takes every element of the list separately and prints it
```

```python
a , b , *c  = packed_tuple
print(a)
print(b)
print(c) ## Here it takes the object as a whole and prints it
```

```python
lsi = [1,2,3,4,5]
print(*lsi)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

#Union 
union_set = set1.union(set2)
union_set

#Intersection
intersection_set = set1.intersection(set2)
intersection_set

#Difference
difference_set = set1.difference(set2)
difference_set

#Symmetric Difference
symmetric_set = set1.symmetric_difference(set2)
symmetric_set

set1.intersection_update(set2) # It takes the common elements between both and updates it to set1 
print(set1)
```
