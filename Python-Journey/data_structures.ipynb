# Packing and Unpacking of tuple

```python
packed_tuple = 1, "Hello" , 3.14, 2 ,3 , 5 ,6,7,8  ## Packing
print(packed_tuple)

```

```python
a , b , c  = packed_tuple ## Unpacking
print(a)
print(b)
print(c)
```

```python
a , b , *c  = packed_tuple
print(a)
print(b)
print(*c) ## Here it takes every element of the list separately and prints it
```

```python
a , b , *c  = packed_tuple
print(a)
print(b)
print(c) ## Here it takes the object as a whole and prints it
```

```python
lsi = [1,2,3,4,5]
print(*lsi)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

#Union 
union_set = set1.union(set2)
union_set

#Intersection
intersection_set = set1.intersection(set2)
intersection_set

#Difference
difference_set = set1.difference(set2)
difference_set

#Symmetric Difference
symmetric_set = set1.symmetric_difference(set2)
symmetric_set

set1.intersection_update(set2) # It takes the common elements between both and updates it to set1 
print(set1)
```

```python
d1 = {"name":"Pranav","Age":22,"marks":100} ## Values are immutable
d2 = d1.copy()
print(d1)
print(d2)
print(d1 is d2)
```

```python
d3 = {"Marks":[10,20,30],"Age":22} ## When the values are mutable , changing in one affects the other 
d4 = d3.copy()
d3["Marks"].append(100)
print(d3)
print(d4)
```

```python
d3['Marks'] = [10]
print(d3)
print(d4)
```

```python
numbers = [1,2,2,2,3,3,3,4,4,4,4]
freq = {}

for x in numbers:
    if(x not in freq):
        freq[x] = 1
    else:
        freq[x]+=1
print(freq)
```
