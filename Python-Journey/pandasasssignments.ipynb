```python
import pandas as pd 
import numpy as np 

def double(x):
    return x*2
df3 = pd.DataFrame(np.random.randint(1,100,size = (5,3)),columns = ['A','B','C'])
df3=df3.apply(double)


def summ_col(col):
    return col.sum()

df3['Total']=df3.apply(summ_col,axis=1)
df3
```
