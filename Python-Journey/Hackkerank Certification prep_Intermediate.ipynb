```python
def most_frequent(nums):
    freq={}

    for i in nums:
        if i in freq:
            freq[i]+=1
        else:
            freq[i]=1
    most_freq = sorted(freq,key=freq.get,reverse=True)
    return most_freq[0]
    

lst = [1,2,3,4,5,5,5,5,57,88,9,8]
most_frequent(lst)
```

```python
def unique_elements(nums):
    unique=[]
    freq={}
    for i in nums:
        if(i in freq):
            freq[i]+=1
        else:
            freq[i]=1 
    
    for j in freq.keys():
        if(freq.get(j)==1):
            unique.append(j)
    return unique


lst = [1,2,3,4,5,5,5,5,57,88,9,8]
print(unique_elements(lst))

```

```python
from collections import Counter 

def first_non_repeating(s):
    freq = Counter(s)
    for i in s:
        if freq[i]==1:
            return i
```

```python
def group_student(lst):
    result = {}
    for i in lst:
        if i[1] in result:
            result.get(i[1]).append(i[0])
        else:
            result[i[1]] = []
            result[i[1]].append(i[0])
    return result 


students = [
    ("Alice", "Math"),
    ("Bob", "Physics"),
    ("Charlie", "Math"),
    ("David", "Physics"),
    ("Eva", "Chemistry")
]
print(group_student(students))

```

```python
from collections import deque

class Queue:
    def __init__(self):
        self.d = deque([])
    
    def enqueue(self,value):
        self.d.append(value)
    def dequeue(self):
        if(len(self.d)==0):
            print("Queue is empty")
        else:
            print(self.d.popleft())
    def length(self):
        return len(self.d)
    
        
        
q = Queue()
q.enqueue(10)
print(q.length())
q.dequeue()
print(q)
print(q.length())
q.enqueue(10)
q.enqueue(20)
print(q.d)
```
