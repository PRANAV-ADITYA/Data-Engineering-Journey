```python
class Bank:
    def __init__(self,owner , balance):
        self.owner = owner 
        self.balance = balance
    
    def deposit(self,amount):
        self.balance+=amount 
        print(f"{amount} is deposited.New balance is {self.balance}")
    
    def withdraw(self,amount):
        if(amount>self.balance):
            print(f"{amount} is not possible as it is more than ur balance which is of {self.balance}")
        elif(self.balance<=0):
            print(f"{amount} is not possible to be drawn as u have less than what ur asking for!")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn from ur account and ur new balance is {self.balance}")
    
    def get_balance(self):
        return self.balance 


# Create an account 
account = Bank("Pranav",34500)
print(account.balance)
account.withdraw(1000000)
    
```

```python
class Car:
    def __init__(self,windows , doors,enginetype):
        self.windows = windows 
        self.doors = doors
        self.enginetype = enginetype
    
    def drive(self):
        print(f"The person will drive the {self.enginetype} car")

class Tesla(Car):
    def __init__(self,*args,is_selfDriving):
        Car.__init__(self,*args)
        self.is_selfDriving = is_selfDriving
    
    def selfDriving(self):
        print(f"Tesla supports {self.is_selfDriving}")
        super().drive()

    def drive(self):
        print("This is from the tesla child class (drive method)")
    

c1 = Tesla("Kuch bhi","Yaar","Matlab whatever ",is_selfDriving="nah")
c1.selfDriving()

c1.drive()
```

# Encapsulation

```python
class Person:
    def __init__(self,name,age,gender):
        self.__name = name # Private variable
        self._age = age # Protected variable
        self.gender = gender 



def get_age(person):
    return person._age

person = Person("Pranav",22,"male")

print(person._age)
```

```python
from abc import ABC , abstractmethod 

class Vehicle(ABC):
    @staticmethod
    def drive():
        print("Car has started to drive")
    
    @abstractmethod
    def start_engine(self):
        pass 

class Cars(Vehicle):
    def start_engine(self):
        print("Car engine started")
    
def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()


car = Cars()
operate_vehicle(car)
Vehicle.drive()
```
