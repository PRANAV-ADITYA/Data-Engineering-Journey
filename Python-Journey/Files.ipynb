# Text File(File Modes)

```python
with open('example.txt','r') as file: # Read mode(text)
    content = file.read() # Reading the entire file at once
    print(content)
```

```python
# Read a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) # It removes whitespace from the beginning and the end(Because of which we dont see the blank lines in between)
```

```python
# Writing to a file (It overwrites the exisiting contents in the file)
with open('example.txt','w') as file:
    file.write('Hello World!\n')
    file.write('This is a new line.')
```

```python
# Writing to a file( It appends new line in the file)(For a single line)
with open('example.txt','a') as file:
    file.write("This line got added due to append mode\n")
```

```python
# It writes new lines without replacing exisiting content in the file(List of lines)
lines = ['pranav\n','aditya\n','gunda\n','kuchbhi\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
with open('example.txt','w+') as file:
    file.write('Hello World\n')
    file.write('This is a new line which is written by this w+ mode \n')

    file.seek(0)

    content = file.read()
    print(content)
```

# Binary Files (File Modes)

```python
# Writing to a binary file

data = b'\x00\x01\x02\x03\x04'

with open('example.bin','wb') as file:
    file.write(data)
```

```python
# Reading from a binary file
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
    

  
```

# Joining Paths

```python
import os
folder_name = 'Folder'
file_name = 'file.txt'
full_path = os.path.join(os.getcwd(),folder_name,file_name)
print(full_path)
```

```python
path = 'example1.txtt'

if os.path.exists(path):
    print(f"{path} exists")
else:
    print(f"{path} does not exist")


if(os.path.isfile(path)):
    print(f"The path {path} is a file")
elif os.path.isdir(path):
    print(f"The path {path} is a directory")
else:
    print(f"The path {path} is neither a file nor a dir")

```
