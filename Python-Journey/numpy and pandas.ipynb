# Numpy


```python
import numpy as np
```



Lists and tuples have a defined order, so NumPy naturally converts each element into an array element. Sets don't have a guaranteed order, so NumPy deliberately treats the entire set as a single Python object (dtype=object) instead of automatically creating a numeric array from its elements.

```python
# Create array using numpy
arr1 = np.array({1,2,3,4,5})
arr1
```

```python
arr2 = np.array([1,2,3,4,5])
arr2
```

```python
np.eye(3) # Used to create an identity matrix (it gives a 2D array)
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print("Shape :" , arr.shape) # Shape of the array
print("No of dimensions:",arr.ndim) # How many dimensions is that array of ?
print("Size:",arr.size) # No of elements in the array
print("Data type:",arr.dtype) #what is the datatype fo that array elements
print("Item size:",arr.itemsize) # How many bytes are being utilised to create the array?
```

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([1,3,4,5,4])
print("Addition:" , arr1+arr2)
print("Substraction:",arr1-arr2)
print("Multiplication:",arr1*arr2)
print("Divison:",arr1/arr2)
```

```python
arr = np.array([2,3,4,5])
print(np.sqrt(arr))
print(np.exp(arr))
print(np.sin(arr))
print(np.log1p(arr))
print(np.log(arr))
```

```python
data = np.array([1,2,3,4,5,6])
mean = np.mean(data)
std_dev = np.std(data)
variance = np.var(data)
median = np.median(data)


normalised_data = (data-mean)/std_dev
print(normalised_data)
print(median)
print(variance)
print(std_dev)
```

```python
# Mergining and Joining DataFrames 
df1 = pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'],'Value2':[4,5,6]})

```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='Key',how="inner")
```

```python
pd.merge(df1,df2,on = "Key",how="outer")
```

```python
pd.merge(df1,df2,on = "Key",how="left")
```

```python
pd.merge(df1,df2,on = "Key",how='right')
```
