```python
import math
def main_square(func):

    def sub_square():
        print("This is the first print statment of the sub_square")
        print(func(25))
        print("This is the second print statement of the sub-square")
    return sub_square()

main_square(math.sqrt)

```

# Decorator
The decorator effectively replaces the function we normally call with a wrapper function that adds extra behavior, while the wrapper still calls the original function internally and preserves its result.

```python
def my_decorator(func):
    def wrapper():
        print("This is something before")
        func()
        print("This is something after")
    return wrapper

```

```python
@my_decorator
def say_hello():
    print("Hello !")

say_hello()
```

```python
def my_decorator1(func):
    def wrapper():
        print("This is something before")
        func()
        print("This is something after")
    return wrapper()

@my_decorator1
def say_hello():
    print("Hello !")

say_hello()
```

```python
def repeat(func):
    def wrapper(msg,n):
        for i in range(n):
            print(msg)
    return wrapper

@repeat
def message(msg):
    print(msg)


message("Hey Pranav",3)
```

```python
def exclaim(func):
    def wrapper1():
        result = func()
        return result+"!"
    return wrapper1

def uppercase(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper

@exclaim
@uppercase
def greeting():
    return "welcome pranav aditya , ur doing fine in python nowadays"


print(greeting())
```

```python
def singleton(cls):
    instances = {}
    def wrapper(*args,**kwargs):
        if cls not in instances:
            instances[cls] = cls(*args,**kwargs)
        return instances[cls]
    return wrapper
            


@singleton
class DatabaseConnection:
    def __init__(self):
        print("The database connection has been established")


w1 = DatabaseConnection()
w2 = DatabaseConnection()
print(w1 is w2)
```
