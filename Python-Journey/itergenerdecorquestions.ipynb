```python

def uppercase(func):
    def wrapper(*args,**kwargs):
        result = func(*args,**kwargs)
        return result.upper()
    return wrapper


class ReverseString:
    def __init__(self,stri):
        self.stri  = stri
    
    @uppercase
    def rev(self):
        words = self.stri.split()
        words.reverse()
        return " ".join(words)
r1 = ReverseString("i am pranav aditya")
print(r1.rev())
```

```python
def singleton(cls):
    instances = {}
    def wrapper(*args,**kwargs):
        if cls not in instances:
            instances[cls] = cls(*args,**kwargs)
            return instances[cls]
    return wrapper
            



class DatabaseConnection:
    print("The database connection has been established")


w1 = DatabaseConnection()
w1

```

```python
import time 



def time_it(func):
    def wrapper(*args,**kwargs):
        start  = time.perf_counter()
        result = func(*args,**kwargs)
        end = time.perf_counter()
        print(f"The execution time is {end-start}")
        return result
    return wrapper



@time_it
def factorial(n):
    result = 1

    for i in range(1,n+1):
        result*=i 
    return result 

print(factorial(5))
# factorial = time_it(factorial)
# factorial(5) ---> wrapper(5)
```

```python
def even_nums(n):
    for i in range(n):
        if(i%2==0):
            yield i 
    
def even_squares(n):
    for i in n :
        yield i**2

f = even_squares(even_nums(10))
for i in f:
    print(i)
```

```python
def squares(n):
    for i in range(n+1):
        yield i**2 

s = squares(10)
for i in s:
    print(i)
```

```python
def fibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        yield a 
        c = a + b
        a = b 
        b = c 

f = fibonacci(5)
for i in f: 
    print(i)
```

```python
class MyRange:
    def __init__(self,start,end,step=2):
        self.start = start 
        self.end = end 
        self.step = step 
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if(self.start>=self.end):
            raise StopIteration
        current = self.start
        self.start+=self.step
        return current

for i in MyRange(1,6,2): 
        print(i)
```

```python
class Countdown:
    def __init__(self,value):
        self.value=value
    
    def __iter__(self):
        return self
    def __next__(self):
        if(self.value<=0):
            raise StopIteration

        self.value-=1
        return self.value

c1 = Countdown(10)
for i in c1:
    print(i)
```
